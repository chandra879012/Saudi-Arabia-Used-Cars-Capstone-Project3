## **1.Business Problem Understanding**

###  Bussines Context
Pasar mobil bekas di Arab Saudi sangat kompetitif, dengan banyak penjual dan pembeli yang bertransaksi setiap hari. Menentukan harga mobil bekas dengan akurat sangat penting bagi penjual, yang ingin memaksimalkan keuntungan, dan pembeli, yang mencari nilai yang adil untuk uang mereka. Kesalahan dalam penetapan harga dapat menyebabkan siklus penjualan yang berkepanjangan, biaya penyimpanan inventaris, atau hilangnya peluang pendapatan.

Dalam pasar ini, memahami faktor-faktor yang mempengaruhi harga mobil dapat membantu dealer, penjual individu, dan pasar mobil online menetapkan harga yang kompetitif, menarik pembeli, dan mempercepat penjualan. Model penetapan harga yang akurat juga dapat membantu pembeli membuat keputusan pembelian yang lebih tepat.

###  Problem Statement
Proses penetapan harga mobil bekas saat ini sering kali didasarkan pada intuisi, data historis, atau aturan sederhana, yang mungkin tidak sepenuhnya mempertimbangkan berbagai faktor yang memengaruhi harga mobil. Hal ini dapat menyebabkan kesalahan penetapan harga, yang dapat menghalangi calon pembeli atau mengurangi margin keuntungan bagi penjual.

Tantangannya adalah mengembangkan model berbasis data yang dapat memprediksi harga jual mobil bekas berdasarkan berbagai fitur seperti jenis mobil, merek, tahun, jarak tempuh, ukuran mesin, dan fitur tambahan. Model ini akan memungkinkan strategi penetapan harga yang lebih akurat dan kompetitif.

###  Goals
Tujuan utama adalah mengembangkan model prediksi yang dapat memperkirakan harga jual mobil bekas di Arab Saudi dengan akurat berdasarkan data yang tersedia. ## Analytic Approach

Langkah yang perlu dilakukan adalah menganalisis seluruh data untuk mengidentifikasi pola dari fitur-fitur yang ada dan perbedaan antara satu mobil dengan yang lain. Selanjutnya, akan dibangun sebuah model regresi yang bertujuan untuk membantu perusahaan dalam menyediakan alat prediksi harga mobil bekas. Pemilihan model regresi akan didasarkan pada evaluasi metrik terbaik sehingga dapat ditentukan model machine learning final yang paling optimal.

## Metric Evaluation

Pada model yang akan dibuat, saat pembersihan data outliers, hanya data outliers yang bernilai ekstrem saja yang akan dihapus. Oleh karena itu, data tetap akan memiliki sejumlah outliers yang didiamkan. Hal ini membuat pada model ini metrik evaluasi yang akan digunakan adalah metrik-metrik yang tidak sensitif terhadap outliers. Metrik seperti MSE (Mean Squared Error), RMSE (Root Mean Squared Error), dan RMSPE (Root Mean Squared Percentage Error) tidak cocok digunakan pada data ini karena sangat sensitif terhadap data outliers. Metrik evaluasi yang akan digunakan adalah:

- **Mean Absolute Error (MAE):**  
  MAE mengukur rata-rata kesalahan absolut antara nilai yang diprediksi dan nilai aktual. Metrik ini tidak memberikan bobot lebih pada outliers, sehingga lebih cocok untuk data yang memiliki outliers yang dibiarkan.

- **Root Mean Squared Logarithmic Error (RMSLE):**  
  RMSLE mengukur kesalahan antara prediksi dan nilai aktual dengan mempertimbangkan skala logaritmik. RMSLE menghitung rata-rata kesalahan logaritmik antara prediksi dan nilai aktual, kemudian mengambil akar kuadratnya.

- **R-squared (R²):**  
  R² mengukur proporsi varians dalam variabel dependen yang dapat diprediksi dari variabel independen. Meskipun R² dapat dipengaruhi oleh outliers, dalam konteks ini, R² digunakan untuk mengukur seberapa baik fitur-fitur yang ada menjelaskan variabilitas dalam data.

- **Mean Absolute Percentage Error (MAPE):**  
  MAPE mengukur kesalahan absolut dalam bentuk persentase. Ini memberikan pandangan tentang kesalahan relatif dalam konteks nilai yang diprediksi, yang lebih mudah diinterpretasikan dan juga tidak terlalu sensitif terhadap outliers.

- **Cek Outlier Menggunakan IQR dan membuat batasan untuk outlier**

```python
# Membuat fungsi untuk mengecek outliers
def find_anomalies(data):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    limit = iqr*1.5
    print(f'IQR: {iqr}')

    limit_bawah = q1 - limit
    limit_atas = q3 + limit
    print(f'limit_bawah: {limit_bawah}')
    print(f'limit_atas: {limit_atas}')
```

- **Cek variance inflation factor(VIF)**

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

dataSAUC = df.select_dtypes(include=['number'])
dataSAUC = dataSAUC.drop(['Price'], axis=1)

def calc_vif(X):
    vif = pd.DataFrame()
    vif["Variable"] = X.columns
    vif["VIF"] = [variance_inflation_factor(X.values, i) for i in range(dataSAUC.shape[1])]
    return(vif)

calc_vif(dataSAUC)
```

 **Penjelasan VIF (Variance Inflation Factor)**

Variance Inflation Factor (VIF) adalah ukuran untuk mengidentifikasi multikolinearitas antara variabel independen dalam sebuah model regresi. Nilai VIF yang lebih tinggi menunjukkan adanya multikolinearitas yang lebih kuat.

- **Choose a Benchmark Model**

**Benchmarking** dalam konteks *regresi machine learning* adalah proses evaluasi dan perbandingan kinerja model-model regresi yang berbeda untuk menentukan model mana yang paling efektif dalam memprediksi variabel target berdasarkan data yang tersedia. Tujuan utama dari benchmarking adalah untuk memilih model yang memberikan prediksi terbaik dengan mempertimbangkan berbagai metrik kinerja. Langkah-langkah umum dalam benchmarking model regresi meliputi pemilihan model, pembagian data menjadi set pelatihan dan pengujian, pelatihan model, evaluasi kinerja menggunakan metrik seperti Root Mean Squared Logarithmic Error (RMSLE), Mean Absolute Percentage Error (MAPE), Mean Absolute Error (MAE), atau R-squared (R²), dan akhirnya perbandingan hasil evaluasi untuk menentukan model terbaik. Benchmarking memastikan bahwa model yang dipilih adalah yang paling sesuai dengan karakteristik data dan tujuan analisis, memberikan hasil yang akurat dan dapat diandalkan.


**Pemilihan Model:** Memilih beberapa algoritma regresi yang akan dibandingkan, seperti Linear Regression, Decision Tree Regression, Random Forest Regression, KNN Regression, Extreme Gradient Boosting Regressor.


```python
# Define algoritma yang digunakan
lr = LinearRegression()
knn = KNeighborsRegressor()
dt = DecisionTreeRegressor(random_state=1)
rf = RandomForestRegressor(random_state=1)
xgb = XGBRegressor(random_state=1)

# Pemodelan dilakukan dalam skala logaritmik, namun kemudian di-inverse kembali untuk interpretasi
log_lr = TransformedTargetRegressor(lr, func=np.log, inverse_func=np.exp)
log_knn = TransformedTargetRegressor(knn, func=np.log, inverse_func=np.exp)
log_dt = TransformedTargetRegressor(dt, func=np.log, inverse_func=np.exp)
log_rf = TransformedTargetRegressor(rf, func=np.log, inverse_func=np.exp)
log_xgb = TransformedTargetRegressor(xgb, func=np.log, inverse_func=np.exp)

# Scaling data
scaler = StandardScaler()

# Kandidat algoritma
models = [log_lr, log_knn, log_dt, log_rf, log_xgb]

score_rmse = []
nilai_mean_rmse = []
nilai_std_rmse = []

score_mae = []
nilai_mean_mae = []
nilai_std_mae = []

score_mape = []
nilai_mean_mape = []
nilai_std_mape = []

score_rmsle = []
nilai_mean_rmsle = []
nilai_std_rmsle = []

score_r2 = []
nilai_mean_r2 = []
nilai_std_r2 = []

# Mencari algoritma terbaik berdasarkan tiap metrics
for i in models:
    
    crossval = KFold(n_splits=5, shuffle=True, random_state=1)

    estimator = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', i)
    ])

    # RMSE
    model_cv_rmse = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_root_mean_squared_error', 
        error_score='raise'
    )
    print(model_cv_rmse, i)
    score_rmse.append(model_cv_rmse)
    nilai_mean_rmse.append(model_cv_rmse.mean())
    nilai_std_rmse.append(model_cv_rmse.std())

    # MAE
    model_cv_mae = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_error', 
        error_score='raise'
    )
    print(model_cv_mae, i)
    score_mae.append(model_cv_mae)
    nilai_mean_mae.append(model_cv_mae.mean())
    nilai_std_mae.append(model_cv_mae.std())

    # MAPE
    model_cv_mape = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_absolute_percentage_error', 
        error_score='raise'
    )
    print(model_cv_mape, i)
    score_mape.append(model_cv_mape)
    nilai_mean_mape.append(model_cv_mape.mean())
    nilai_std_mape.append(model_cv_mape.std())

    # RMSLE
    model_cv_rmsle = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='neg_mean_squared_log_error',  # Scikit-learn tidak memiliki RMSLE, jadi gunakan MSLE dan akarkan hasilnya
        error_score='raise'
    )
    print(model_cv_rmsle, i)
    score_rmsle.append(model_cv_rmsle)
    nilai_mean_rmsle.append((-model_cv_rmsle.mean()) ** 0.5)  # Mengubah MSLE menjadi RMSLE dengan mengakarkan
    nilai_std_rmsle.append(model_cv_rmsle.std())

    # R-squared (R2)
    model_cv_r2 = cross_val_score(
        estimator, 
        X_train, 
        y_train, 
        cv=crossval, 
        scoring='r2', 
        error_score='raise'
    )
    print(model_cv_r2, i)
    score_r2.append(model_cv_r2)
    nilai_mean_r2.append(model_cv_r2.mean())
    nilai_std_r2.append(model_cv_r2.std())
```

**Hyperparameter Tuning**

Pada tahap ini, Saya akan melakukan *hyperparameter tuning* untuk menemukan kombinasi parameter terbaik pada model terpilih, yaitu *Extreme Gradient Boost*. Mengingat bahwa *Extreme Gradient Boost* memiliki banyak parameter, mengganti seluruh parameter bawaan dapat meningkatkan risiko *overfitting*. Oleh karena itu, Saya hanya akan memodifikasi beberapa parameter utama:

Dengan menyesuaikan parameter-parameter ini, diharapkan model *Extreme Gradient Boost* yang dihasilkan akan memiliki keseimbangan yang lebih baik antara akurasi dan generalisasi, sehingga mengurangi risiko *overfitting* dan menghasilkan prediksi yang lebih andal.


```python
# Kedalaman pohon
max_depth = list(np.arange(1, 11))

# Learning rate
learning_rate = list(np.arange(1, 100)/100)

# Jumlah pohon
n_estimators = list(np.arange(100, 201))

# Jumlah baris tiap pohon (% dari total baris train set)
subsample = list(np.arange(2, 10)/10)

# Gamma (min_impurity_decrease)
gamma = list(np.arange(1, 11)) # Semakin besar nilainya, semakin konservatif/simpel modelnya

# Jumlah feature yang digunakan untuk tiap pohon (% dari total kolom train set)
colsample_bytree = list(np.arange(1, 10)/10)

# Alpha (regularization)
reg_alpha = list(np.logspace(-3, 1, 10)) # Semakin besar nilainya, semakin konservatif/simpel modelnya


# Hyperparam space XGboost
hyperparam_space_xgb = {
    'model__max_depth': max_depth, 
    'model__learning_rate': learning_rate,
    'model__n_estimators': n_estimators,
    'model__subsample': subsample,
    'model__gamma': gamma,
    'model__colsample_bytree': colsample_bytree,
    'model__reg_alpha': reg_alpha
}
```

```python
# Benchmark model dengan hyperparameter tuning
xgb = XGBRegressor(random_state=1, verbosity=0)

# Membuat algorithm chains
estimator_xgb = Pipeline([
        ('preprocessing', transformer),
        ('scaler', scaler),
        ('model', xgb)
        ])

crossval = KFold(n_splits=5, shuffle=True, random_state=1)

# Hyperparameter tuning
random_xgb = RandomizedSearchCV(
    estimator_xgb, 
    param_distributions = hyperparam_space_xgb,
    n_iter = 50,
    cv = crossval, 
    scoring = ['neg_root_mean_squared_error', 'neg_mean_absolute_error', 'neg_mean_absolute_percentage_error'], 
    n_jobs = -1,
    refit = 'neg_root_mean_squared_error', # Hanya bisa memilih salah stau metric untuk optimisasi
    random_state = 1  
)
```

```python
#Fitting data train untuk mencari parameter terbaik
random_xgb.fit(X_train, y_train)
```

Pipeline di atas digunakan dalam model machine learning untuk melakukan preprocessing data dan prediksi menggunakan **XGBoost Regressor (XGBRegressor)**. Berikut adalah cara kerjanya:

1. **Preprocessing dengan ColumnTransformer**:
   - **OneHotEncoder**: Mengubah variabel kategorikal menjadi format one-hot encoding.
   - **BinaryEncoder**: Mengubah variabel kategorikal biner ke dalam representasi biner.
   - **RobustScaler**: Menerapkan skala robust pada fitur yang rentan terhadap outliers, dengan menghilangkan median dan menskalakan data berdasarkan interquartile range (IQR).
   - **Remainder (Passthrough)**: Fitur yang tidak diproses oleh transformer lain akan diteruskan langsung tanpa perubahan.

2. **StandardScaler**:
   - Setelah preprocessing, semua fitur distandardisasi dengan **StandardScaler** untuk menormalkan fitur sehingga memiliki mean 0 dan standar deviasi 1.

3. **XGBRegressor**:
   - Model **XGBoost Regressor** digunakan untuk memprediksi target berdasarkan fitur yang telah diproses.

4. **RandomizedSearchCV**:
   - Pipeline ini dioptimalkan menggunakan **RandomizedSearchCV**, yang mencari kombinasi terbaik dari hyperparameter untuk model XGBRegressor melalui pencarian acak dalam ruang hyperparameter yang telah ditentukan.

Dengan pipeline ini, keseluruhan proses preprocessing dan model training/prediction terintegrasi dalam satu langkah, mempermudah replikasi dan tuning model.
